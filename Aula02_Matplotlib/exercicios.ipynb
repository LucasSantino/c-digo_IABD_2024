```python
import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# EXERCÍCIO 1

df['data']=pd.to_datetime(df['data'])
plt.figure(figsize=(10,6))
plt.plot(df['data'],df['humidade'])
plt.title('Umidade x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Uumidade')
plt.legend('Umidade',loc='lower right')
```

```python
# EXERCÍCIO 2

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['velocidade do vento'])
plt.title('Velocidade do vento x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Velocidade do vento')
plt.legend('Velocidade do vento', loc='upper right')
```

```python
# EXERCÍCIO 3

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['humidade'])
plt.title('Umidade x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Umidade')
plt.legend(['Umidade'], loc='lower right')

# Anotação no gráfico
plt.text(df['data'].iloc[0], df['humidade'].iloc[0] - 1, 'Início do período', 
         fontsize=12, color='red', verticalalignment='top')
```

```python
# EXERCÍCIO 4

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['direção do vento'])
plt.title('Direção do Vento x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Direção do Vento')
plt.legend(['Direção do Vento'], loc='upper right')

# Anotação no gráfico
plt.text(df['data'].iloc[10], df['direção do vento'].iloc[10] + 5, 'Mudança significativa', 
         fontsize=12, color='blue', verticalalignment='bottom')
```

```python
# EXERCÍCIO 5

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['pressão'])
plt.title('Pressão x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Pressão')
plt.legend(['Pressão'], loc='upper right')

# Anotação no gráfico
plt.text(df['data'].iloc[5], df['pressão'].iloc[5] + 1, 'Aumento de pressão', 
         fontsize=12, color='red', verticalalignment='bottom')
```

```python
# EXERCÍCIO 9

df_2016 = df[(df['data'] >= '2016-05-01') & (df['data'] <= '2016-08-31')]
df_2016_full = df[(df['data'] >= '2016-01-01') & (df['data'] <= '2016-12-31')]

fig = plt.figure(figsize=(12,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df_2016['data'], df_2016['temperatura'], color='b')
eixo.grid(True)
eixo.set_xlim([datetime.datetime(2016,5,1), datetime.datetime(2016,8,31)])
eixo.set_title('Temperatura de Maio a Agosto de 2016', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['temperatura'], loc='best')

eixo2 = fig.add_axes([0.35,0.50,0.5,0.3])
eixo2.plot(df_2016_full['data'], df_2016_full['temperatura'], color='r')
eixo2.set_xlim([datetime.datetime(2016,1,1), datetime.datetime(2016,12,31)])
eixo2.tick_params(axis='x', labelsize=8)
eixo2.grid(True)
eixo2.set_title('Temperatura ao longo de 2016', fontsize=15)
eixo2.set_ylabel('Temperatura', fontsize=20)
eixo2.set_xlabel('Ano', fontsize=20)
eixo2.legend(['temperatura'], loc='best', fontsize=10)
```

```python
# EXERCÍCIO 10

df_2017 = df[(df['data'] >= '2017-05-01') & (df['data'] <= '2017-08-31')]
df_2017_umidade = df[(df['data'] >= '2017-05-01') & (df['data'] <= '2017-08-31')]

fig = plt.figure(figsize=(12,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df_2017['data'], df_2017['temperatura'], color='b')
eixo.grid(True)
eixo.set_xlim([datetime.datetime(2017,5,1), datetime.datetime(2017,8,31)])
eixo.set_title('Temperatura de Maio a Agosto de 2017', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.legend(['temperatura'], loc='best')

eixo2 = fig.add_axes([0.35,0.50,0.5,0.3])
eixo2.plot(df_2017_umidade['data'], df_2017_umidade['humidade'], color='r')
eixo2.set_xlim([datetime.datetime(2017,5,1), datetime.datetime(2017,8,31)])
eixo2.tick_params(axis='x', labelsize=8)
eixo2.grid(True)
eixo2.set_title('Umidade de Maio a Agosto de 2017', fontsize=15)
eixo2.set_ylabel('Umidade', fontsize=20)
eixo2.set_xlabel('Ano', fontsize=20)
eixo2.legend(['humidade'], loc='best', fontsize=10)
```
