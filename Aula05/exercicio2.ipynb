```python
import numpy as np # cria a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# Skylearn importa o modelo de regress√£o linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
temperatura = np.array([30,35,40,45,60,100,150]). reshape (-1, 1)#Temperatura do forno
                       
quantidadeCalor = np.array([100,200,300,400,500,600,700]) # quantidade de calor
#cria e treiba o modelo de regress√£o linear
modelo = LinearRegression() #cria uma variavel chamada modelo
modelo.fit(temperatura, quantidadeCalor) # cria um modelo com base nos dados de entrada
```

```python
# Dados de teste
calorTemperatura = np.array([5, 20, 7,30,50, 40, 70]).reshape(-1, 1)  # temperatura
# Fazer previs√µes com base na temperatura
nivel_modelo= modelo.predict(calorTemperatura)
```

```python
# Imprimir as previs√µes
for i in range(len(calorTemperatura)): # la√ßo de repeti√ß√£o para exibir os dados
    # de saida
    print(f"Calor: {calorTemperatura[i][0]} - N√≠vel previsto: {nivel_modelo[i]} temperatura")
# gr√°fico de pontos - scatter
plt.scatter(calorTemperatura, nivel_modelo, color='blue', label='Dados de Treinamento') 
# gr√°fico de linhas 
plt.plot(temperatura,quantidadeCalor, color  = 'red',label='Dados de teste')
plt.xlabel('Temperatura')
plt.ylabel('Quantidade de calor ')
plt.title('Rela√ß√£o entre Temperatura e Quantidade de calor')
plt.legend()
plt.grid(True)
plt.show()
```
