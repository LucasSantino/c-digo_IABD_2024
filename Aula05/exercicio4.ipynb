```python
import numpy as np # cria a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# Skylearn importa o modelo de regress√£o linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
Renda = np.array([1200,5000,8000,10000,30000]). reshape (-1, 1)#Temperatura do forno
Gastos= np.array([750,1500,4000,6000,18000]) # quantidade de calor
#cria e treiba o modelo de regress√£o linear
modelo = LinearRegression() #cria uma variavel chamada modelo
modelo.fit(Renda, Gastos) # cria um modelo com base nos dados de entrada
```

```python
# Dados de teste
GastosClientes= np.array([1400, 3000, 4580,5500,1800, 20000, 31000]).reshape(-1, 1)  
# Fazer previs√µes com base na temperatura
RendaSalario_Modelo= modelo.predict(GastosClientes)
```

```python
# Imprimir as previs√µes
for i in range(len(GastosClientes)): # la√ßo de repeti√ß√£o para exibir os dados
    # de saida
    print(f"Area: {GastosClientes[i][0]} - N√≠vel previsto: {RendaSalario_Modelo[i]} temperatura")
# gr√°fico de pontos - scatter
plt.scatter(GastosClientes, RendaSalario_Modelo, color='blue', label='Dados de Treinamento') 
# gr√°fico de linhas 
plt.plot(Renda,Gastos, color  = 'red',label='Dados de teste')
plt.xlabel('Gastos Previsto')
plt.ylabel('Renda ')
plt.title('Previs√£o sos gastos dos clientes com base no sal√°rio.')
plt.legend()
plt.grid(True)
plt.show()
```
