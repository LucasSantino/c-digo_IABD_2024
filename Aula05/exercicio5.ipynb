```python
import numpy as np # cria a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# Skylearn importa o modelo de regressão linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
Consumo = np.array([50, 140, 30, 75, 123]). reshape (-1, 1)#Temperatura do forno
Demanda= np.array([100, 200, 80, 100, 160]) # quantidade de calor
#cria e treiba o modelo de regressão linear
modelo = LinearRegression() #cria uma variavel chamada modelo
modelo.fit(Consumo, Demanda) # cria um modelo com base nos dados de entrada
```

```python
# Dados de teste
ConsumoMensalClientes= np.array([50, 70, 140,150,60]).reshape(-1, 1)  
# Fazer previsões com base na demanda de energia
DemandaEnergia= modelo.predict(ConsumoMensalClientes)
```

```python
# Imprimir as previsões
for i in range(len(ConsumoMensalClientes)): # laço de repetição para exibir os dados
    # de saida
    print(f"Area: {ConsumoMensalClientes[i][0]} - Nível previsto: {DemandaEnergia[i]} energia")
# gráfico de pontos - scatter
plt.scatter(ConsumoMensalClientes, DemandaEnergia, color='blue', label='Dados de Treinamento') 
# gráfico de linhas 
plt.plot(Consumo,Demanda, color  = 'red',label='Dados de teste')
plt.xlabel('Consumo Mensal Previsto pelos Clientes')
plt.ylabel('Demanda de Energia ')
plt.title('Previsão sda demanda de energia com base no consumo dos meses anteriores pelos clientes.')
plt.legend()
plt.grid(True)
plt.show()
```
