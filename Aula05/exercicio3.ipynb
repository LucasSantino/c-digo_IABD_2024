```python
import numpy as np # cria a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# Skylearn importa o modelo de regressão linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
Area = np.array([50,100,120,300] ). reshape (-1, 1)# area  do apartamento               
Valor = np.array([180.000,300.000,375.000,600.000]) # valor imovel
#cria e treiba o modelo de regressão linear
modelo = LinearRegression() #cria uma variavel chamada modelo
modelo.fit(Area, Valor) # cria um modelo com base nos dados de entrada
```

```python
# Dados de teste
AreaImovel = np.array([50, 200, 280,300,584, 689, 752]).reshape(-1, 1)  
# Fazer previsões com base na temperatura
ModeloImovel= modelo.predict(AreaImovel)
```

```python
# Imprimir as previsões
for i in range(len(AreaImovel)): # laço de repetição para exibir os dados
    # de saida
    print(f"Area: {AreaImovel[i][0]} - Nível previsto: {ModeloImovel[i]} temperatura")
# gráfico de pontos - scatter
plt.scatter(AreaImovel, ModeloImovel, color='blue', label='Dados de Treinamento') 
# gráfico de linhas 
plt.plot(Area,Valor, color  = 'red',label='Dados de teste')
plt.xlabel('Valor Previsto')
plt.ylabel('Area ')
plt.title('Previsão od valor de um imovel com base na área')
plt.legend()
plt.grid(True)
plt.show()
```
