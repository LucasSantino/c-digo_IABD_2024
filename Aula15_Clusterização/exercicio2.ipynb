```python
#Importação das bibliotecas

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Passo 2: Criar um DataFrame com os dados dos vinhos
dados_vinhos = pd.DataFrame({
 'Teor Alcoólico': [3, 4, 5, 6],
 'Acidez': ['muito', 'pouco', 'médio', 'baixo'],
 'pH': [4.3, 2.8, 4.2, 3.9]
})
```

```python
dados_vinhos = pd.get_dummies(dados_vinhos, columns=['Acidez'])
```

```python
# Passo 4: Aplicar o algoritmo K-means
kmeans = KMeans(n_clusters=2) # Definindo 2 clusters
kmeans.fit(dados_vinhos)
```

```python
# Passo 5: Avaliar o desempenho do modelo K-means
sse = kmeans.inertia_
print("SSE:", sse)
```

```python
# Passo 6: Visualizar os resultados da segmentação
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Plotando um gráfico de dispersão dos vinhos com cores diferentes para cada grupo
plt.scatter(dados_vinhos['Teor Alcoólico'], dados_vinhos['pH'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 2], marker='x', color='red')
plt.xlabel('Teor Alcoólico')
plt.ylabel('pH')
plt.title('Segmentação de Vinhos')
plt.show()
```

```python
# Plotando um gráfico de dispersão dos vinhos com cores diferentes para cada grupo
plt.scatter(dados_vinhos['Teor Alcoólico'], dados_vinhos['pH'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 2], marker='x', color='red')
plt.xlabel('Teor Alcoólico')
plt.ylabel('pH')
plt.title('Segmentação de Vinhos')
plt.show()
```

```python
# Passo 7: Analisar os grupos resultantes
dados_vinhos['Grupo'] = labels
grouped_data = dados_vinhos.groupby('Grupo').mean()
print(grouped_data)
```
