```python
#Importando as bibliotecas
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans # importa a biblioteca Kmeans
import matplotlib.pyplot as plt
```

```python
#Criando o conjunto de dados
setores = np.array([1,2,3,4,5,6,7,8])
nº_produtosFabricados = np.array([100,50,15,200,500,1000,375,450])
```

```python
#Juntano as colunas dos dados
dados_produtivos = np.column_stack((setores,nº_produtosFabricados))
```

```python
dados_produtivos
```

```python
# Aplicando o algoritimo Kmeans
kmeans = KMeans(n_clusters=3) # definido o numero de clusters
kmeans.fit(dados_produtivos) #Criando o modelo para realizar o agrupamento
```

```python
#Avaliando a métrica do kmeans
sse = kmeas.inertia_
print("SSE: ",sse)
```

```python
#Visualizando o resultado dos agrupamentos
labels = kmeas.labels_
centroids = kmeas.cluster_centers_
```

```python
# Plotando um gráfico de dispersão dos clientes com cores diferentes para cada grupo
plt.scatter(dados_produtivos[:, 0], dados_produtivos[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
plt.xlabel('Setores')
plt.ylabel('Nº_produtosFabricados')
plt.title('Setores mais produtivos')
plt.show()
```
