```python
import pandas as pd #Importa a biblioteca pandas
import matplotlib.pyplot as plt # Importa a biblioteca Matplotlib
import numpy as np # Importa a biblioteca numpy 
```

```python
#Gerando numeros com a numpy
dados_x = np.linspace(0,10,10) #gera valores com a biblioteca umpy
dados_y = dados_x * dados_x # gera valores multipliando um dado pelo o outro
```

```python
plt.figure(figsize=(10,5), facecolor= 'gray')
Ax = plt.axes() # CRIA OS EIXOS DO GR√ÅFICO
Ax.set_facecolor('white') #CRIA O FUNDO DO GRADICO
plt.plot(dados_x, dados_y, 
         color='red', linewidth=2, 
         marker= 'o', linestyle = 'dashed', markeredgecolor = 'black')

plt.xlabel("Eixo x ", fontsize=12 )  #Adiciona titulo do eixo x

plt.ylabel("Eixo y", fontsize=12) #Adiciona titulo do eixo y

plt.title("Exemplo Matiplotlib", fontsize=12) #Adiciona titulo do grafico
```

```python
plt.rcParams.update({'font.size':15}) #define o tamanho da fonte
plt.figure(figsize=(12,12)) #define o tamanho da figura
linhas = 2 #deine o numero de linhas
colunas = 2 #Define o numero de colunas
plt.subplot(linhas,colunas,1) #define o numero de linhas de linhas e colunas para o subplot do gr√°fico
plt.plot(dados_x, 'r--')
plt.title("grafico 1")

#grafio2

plt.subplot(linhas, colunas, 2)
plt.plot(dados_x*dados_y, 'b--')
plt.title("Grafico 2")

#grafico 3
plt.subplot(linhas, colunas, 3)
plt.plot(dados_x*dados_y, 'c--')
plt.title("Grafico 3")

#Grafico 4
plt.subplot(linhas, colunas, 4)
plt.plot(dados_x*2, 'r--')
plt.title("Graficos 4 ")
plt.savefig("Exemplo subplot.jpg", dpi=300, transparent= True, #Salva a imagem dos graficos
            bbox_inches='tight')
```

```python
# Grafico de barras
figura = plt.figure() #Arazena na variavel  figura a  a figura a se 
grafico = figura.add_axes([0,0,0.7,0.7]) #Cria a grade do gr√©fico
rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'] # cria os rotulos 
grafico.bar(rotulos,dados_x)
plt.title("Quantidade X Categorias")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
#Grafico de barras
figura = plt.figure() #Arazena na variavel  figura a  a figura a se 
grafico = figura.add_axes([0,0,0.7,0.7]) #Cria a grade do gr√©fico
rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'] # cria os rotulos 
grafico.bar(rotulos,dados_x, color='blue')
grafico.bar(rotulos,dados_x*0.1, color='red')
plt.title("Quantidade X Categorias")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```
